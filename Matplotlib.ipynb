```python
data = {
    'a' : np.arange(5),
    'c':np.random.randint(0,5,5),
    'd': np.random.randn(5)
}
data['b'] = data['a'] + 10 * np.random.randn(5)
data['d']= np.abs(data['d']) * 100
print(data)
```

```python
plt.scatter('a','b', c='b', s='d',data=data)
plt.show()
```

```python
names = ['g_a', 'g_b', 'g_c']
values = [1,10,100]
plt.figure(figsize=(9,3))

plt.subplot(131)# (row, column, position)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names,values)
plt.subplot(133)
plt.plot(names,values)
plt.suptitle('Categorical Plotting')
plt.show()
```
